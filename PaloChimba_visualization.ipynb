```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.set(style='whitegrid')
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.distplot(df['Age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.countplot(data=df, y='Occupation')
plt.title('Occupation Distribution')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.boxplot(data=df, y='Sleep Duration')
plt.title('Box Plot of Sleep Duration')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Sleep Duration')
plt.title('Box Plot of Sleep Duration')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.scatterplot(data=df, x='Sleep Duration', y='Quality of Sleep')
plt.title('Sleep Duration vs. Quality of Sleep')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep (1-10)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Physical Activity Level', y='Heart Rate')
plt.title('Physical Activity Level vs. Heart Rate')
plt.xlabel('Physical Activity Level')
plt.ylabel('Resting Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='BMI Category', y='Daily Steps')
plt.title('Daily Steps Across Different BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Daily Steps')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
heatmap_data = df.groupby(['Stress Level', 'Quality of Sleep']).size().unstack()
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Stress Levels vs. Sleep Quality')
plt.xlabel('Sleep Quality')
plt.ylabel('Stress Level')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.pairplot(df[['Age', 'BMI Category', 'Sleep Duration', 'Physical Activity Level', 'Stress Level']])
plt.title('Pair Plot of Key Variables')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.boxplot(x='BMI Category', y='Daily Steps', data=df, palette='Set2')
plt.title('Daily Steps Distribution by BMI Category', fontsize=16)
plt.xlabel('BMI Category', fontsize=14)
plt.ylabel('Daily Steps', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
plt.scatter(df['Physical Activity Level'], df['Heart Rate'], alpha=0.6 , cmap='YlGnBu')
plt.title('Physical Activity Level vs. Heart Rate', fontsize=16)
plt.xlabel('Physical Activity Level', fontsize=14)
plt.ylabel('Resting Heart Rate (bpm)', fontsize=14)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
df.groupby('Age')['Heart Rate'].mean().plot(kind='line', marker='o', color='magenta')
plt.title('Average Heart Rate by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Heart Rate (bpm)', fontsize=14)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Quality of Sleep'], color='magenta')
plt.title('Distribution of Sleep Quality', fontsize=16)
plt.xlabel('Sleep Quality (scale 1-10)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.stripplot(df['Physical Activity Level'], df['Heart Rate'], alpha=0.6, jitter = True)
plt.title('Physical Activity Level vs. Heart Rate', fontsize=16)
plt.xlabel('Physical Activity Level', fontsize=14)
plt.ylabel('Resting Heart Rate (bpm)', fontsize=14)
plt.grid()
plt.show()
```

```python
sns.set(style="whitegrid")

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Stress Level', data=df, inner='quartile')

# Add title and labels
plt.title('Violin Plot of Stress Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Stress Level')
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Age'].value_counts().plot(kind='bar')
plt.title('Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Create a scatter plot
sns.scatterplot(data=df, x='Age', y='BMI Category', hue='Sleep Disorder', style='Sleep Disorder', s=100)

# Customize the plot
plt.title('Sleep Disorder vs BMI and Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('BMI', fontsize=14)
plt.legend( bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Create a count plot for Sleep Disorder by Gender
sns.countplot(data=df, x='Sleep Disorder', hue='Gender', palette='Set2')

# Customize the plot
plt.title('Sleep Disorder Distribution by Gender', fontsize=16)
plt.xlabel('Sleep Disorder', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Gender')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Physical Activity Level'], df['Heart Rate'], marker='o', linestyle='-', color='g')

# Step 3: Adding titles and labels
plt.title('Physical Activity Level vs. Heart Rate')
plt.xlabel('Physical Activity Level (1-10)')
plt.ylabel('Heart Rate (bpm)')
  # Set x-ticks from 1 to 10
plt.grid()
```

```python
sns.set(style='whitegrid')

# Step 3: Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='BMI Category', y='Daily Steps', palette='Set2')

# Step 4: Adding titles and labels
plt.title('Daily Steps Across Different BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Daily Steps')
plt.grid()

# Step 5: Show the plot
plt.show()
```
